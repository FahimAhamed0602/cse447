```python
# generate large prime(128 bits) p and q
import random
import sympy

# 128 bits means the number should be between 2^127 and 2^128
lower_bound = 2**127
upper_bound = 2**128

p = sympy.randprime(lower_bound, upper_bound)
q = sympy.randprime(lower_bound, upper_bound)

# making sure p and q aren't the exact same number just in case
while p == q:
    q = sympy.randprime(lower_bound, upper_bound)

print("p =", p)
print("q =", q)
```

```python
#Calculate n = pq

n = p * q
print("n =", n)
```

```python
#Calculate phi(n)=(p-1)(q-1)

phi = (p - 1) * (q - 1)
print("phi(n) =", phi)
```

```python
#Perform encryption here and generate ciphertext [c = m^e (mod n)] and private key [d*e ≡ 1 (mod(phi(n)))]
e = 11

def egcd(a, b):
  if a == 0:
    return (b, 0, 1)
  else:
    g, y, x = egcd(b % a, a)
    return (g, x - (b // a) * y, y)

# function to find the modular inverse using egcd
def modinv(a, m):
  g, x, y = egcd(a, m)
  if g != 1:
    print("Error: modular inverse does not exist")
    return None
  else:
    return x % m

d = modinv(e, phi)
print("Private key d =", d)

# Encrypting the message: c = m^e (mod n)
c = pow(m, e, n)
print("Ciphertext c =", c)
```

```python
#perform decryption here [m = c^d (mod n)]
decrypted_m = pow(c, d, n)
print("Decrypted m =", decrypted_m)
```

```python
m = decrypted_m
# The lab provided this line to decode the hex back to text
print("Final Output:", bytes.fromhex(hex(m)[2:]).decode('utf-8'))
```
